## Markdown Parsing

**What it does:** splits a Markdown file into sections based on its headings. 

**When to use it:** When your documents are Markdown files (`.md`)

**Steps:**
1. Read the file.
2. Split on heading lines (lines starting with `#`)
3. Store each section with its heading.

```python
sample_md = "./tmp/sample_nb.md"

with open(sample_md, "w") as f:
    f.write(
        "# API Docs\n\n"
        "## Auth\n"
        "Use a Bearer token.\n\n"
        "## Endpoints\n"
        "GET /users returns all users.\n"
    )
```

```python
import re

with open(sample_md) as f:
    data_from_markdown_file = f.read()


sections = re.split(
    r'^(#{1,3}\s+.+)$',
    data_from_markdown_file, 
    flags = re.MULTILINE
)

print(sections)

for i, section in enumerate(sections):
    print(f"SECTION {i+1}:", section.strip())
    
```

```python
for section in sections:
    section = section.strip()

    if not section:
        continue
    if section.startswith("#"):
        heading = section.lstrip("#").strip()
    else:
        print(f"[{heading}] {section}")
```

**What happened:** We split the Markdown content whenever we found a heading line. Each section is stored with its heading name.
