## HTML Parsing with BeautifulSoup

**What it does:** Reads an HTML web page and extracts the useful text, removing navigation bars, scripts, and other noise. 

**When to use it:** When your source documents are web pages. 

**Steps:**
1. Parse the HTML. 
2. Remove unwanted tags (scripts, nav, footer).
3. Get the clean text. 

```python
from bs4 import BeautifulSoup
```

```python
sample_html = """
<html>
<head><title>Cloud Guide</title></head>
<body>
  <nav>Home | About</nav>
  <article>
    <h1>Cloud Computing</h1>
    <p>Cloud provides on-demand computing.</p>
    <h2>Service Types</h2>
    <ul>
      <li>IaaS - Infrastructure</li>
      <li>PaaS - Platform</li>
      <li>SaaS - Software</li>
    </ul>
  </article>
  <footer>Copyright 2024</footer>
  <script>console.log('tracking');</script>
</body>
</html>
"""
```

**What happened:** We defined a small HTML page as a string so we can practice parsing it. 

### Extract Clean Text

```python
soup = BeautifulSoup(sample_html, "html.parser")

for tag in soup(["script", "style", "nav", "footer"]):
    print(tag)
    tag.decompose()
```

```python
text = soup.get_text(separator="\n", strip=True)
print(text)
```

**What happened:** `decompose()` deletes a tag and everything inside it. We removed scripts, styles, navigation, and footer, then printed the remaining text.

### Extract Headings and Links 

```python
soup = BeautifulSoup(sample_html, "html.parser")

heading_tags = soup.find_all(["h1", "h2", "h3"])

print(heading_tags)
```

```python
headings = []

for tag in heading_tags:
    heading_info = {
        "level": tag.name,
        "text": tag.get_text()
    }
    headings.append(heading_info)
```

```python
for heading in headings:
    print(heading)
```

**What happened:** We found every heading tag (h1, h2, h3) in the HTML and printed its level and text. 

## Table Extraction from HTML

**What it does:** Reads an HTML `<table>` tag and converts it to a list of dictionaries. 

**When to use it:** When a web page contains data tables you want to use. 

```python
table_html = """
<table>
  <tr><th>Provider</th><th>Free Tier</th></tr>
  <tr><td>AWS</td><td>12 months</td></tr>
  <tr><td>GCP</td><td>$300 credits</td></tr>
</table>
"""
```

```python
soup = BeautifulSoup(table_html, "html.parser")
table = soup.find("table")

print(table)
```

```python
rows = table.find_all("tr")
print(rows)
```

```python
header_tags = rows[0].find_all("th")

print(header_tags)
header_tags
```

```python
headers = []
for header in header_tags:
    headers.append(header.get_text())

print(headers)
```

```python
data = [] 

for row in rows[1:]:
    cell_tags = row.find_all("td")
    print("CELL TAGS:\n", cell_tags, "\n")

    cells = []
    for td in cell_tags:
        cells.append(td.get_text())
    
    print("CELLS:\n", cells, "\n")


    row_dictionary = {}
    for i in range(len(headers)):
        row_dictionary[headers[i]] = cells[i]
    
    print("ROW DICTIONARY:\n", row_dictionary, "\n")
    data.append(row_dictionary)


print("DATA:\n", data, "\n")
    
```

**What happened:** We read the header row to get column names, then read each data row and combined them into dictionaries.
